## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

```python
# ĐIỀN CODE VÀO ĐÂY
q12 = pd.read_sql(""" SELECT team_api_id, team_long_name, SUM(total_goals) AS total_goals
                        FROM (
                            SELECT m.home_team_api_id AS team_api_id, t.team_long_name, SUM(m.home_team_goal) AS total_goals
                            FROM Team t
                            JOIN Match m ON t.team_api_id = m.home_team_api_id
                            GROUP BY m.home_team_api_id
                            
                            UNION ALL
                            
                            SELECT m.away_team_api_id AS team_api_id, t.team_long_name, SUM(m.away_team_goal) AS total_goals
                            FROM Team t
                            JOIN Match m ON t.team_api_id = m.away_team_api_id
                            GROUP BY m.away_team_api_id
                        ) AS combined
                        GROUP BY team_api_id, team_long_name
                        ORDER BY total_goals DESC
                        ;  """, conn)
q12
```

## Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.

```python
# ĐIỀN CODE VÀO ĐÂY
q13 = pd.read_sql(""" SELECT team_api_id, team_long_name, SUM(match_count) AS match_count
                        FROM (
                            SELECT m.home_team_api_id AS team_api_id, t.team_long_name, COUNT(m.home_team_api_id) AS match_count
                            FROM Team t
                            JOIN Match m ON t.team_api_id = m.home_team_api_id
                            GROUP BY m.home_team_api_id
                            
                            UNION ALL
                            
                            SELECT m.away_team_api_id AS team_api_id, t.team_long_name, COUNT(m.away_team_api_id) AS match_count
                            FROM Team t
                            JOIN Match m ON t.team_api_id = m.away_team_api_id
                            GROUP BY m.away_team_api_id
                        ) AS combined
                        GROUP BY team_api_id, team_long_name
                        ORDER BY match_count DESC
                        ;  """, conn)
q13
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"


```python
# ĐIỀN CODE VÀO ĐÂY
q14 = pd.read_sql(""" SELECT team_api_id, team_long_name,
                            SUM(CASE WHEN result = 'win' THEN 1 ELSE 0 END) AS wins,
                            SUM(CASE WHEN result = 'lose' THEN 1 ELSE 0 END) AS losses,
                            SUM(CASE WHEN result = 'draw' THEN 1 ELSE 0 END) AS draws
                        FROM (
                                SELECT m.home_team_api_id AS team_api_id, t.team_long_name,
                                       CASE
                                           WHEN m.home_team_goal > m.away_team_goal THEN 'win'
                                           WHEN m.home_team_goal < m.away_team_goal THEN 'lose'
                                           ELSE 'draw'
                                       END AS result
                                FROM Match m
                                JOIN Team t ON m.home_team_api_id = t.team_api_id
                                UNION ALL
                                SELECT m.away_team_api_id AS team_api_id, t.team_long_name,
                                       CASE
                                           WHEN m.home_team_goal < m.away_team_goal THEN 'win'
                                           WHEN m.home_team_goal > m.away_team_goal THEN 'lose'
                                           ELSE 'draw'
                                       END AS result
                                FROM Match m
                                JOIN Team t ON m.away_team_api_id = t.team_api_id
                        ) AS combined_results
                        GROUP BY team_api_id, team_long_name
                        ORDER BY team_api_id
                        ;  """, conn)
q14
```

## Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất

```python
# ĐIỀN CODE VÀO ĐÂY
q14 = pd.read_sql(""" SELECT team_api_id, team_long_name,
                            SUM(CASE WHEN result = 'win' THEN 1 ELSE 0 END) AS wins,
                            SUM(CASE WHEN result = 'lose' THEN 1 ELSE 0 END) AS losses,
                            SUM(CASE WHEN result = 'draw' THEN 1 ELSE 0 END) AS draws
                        FROM (
                                SELECT m.home_team_api_id AS team_api_id, t.team_long_name,
                                       CASE
                                           WHEN m.home_team_goal > m.away_team_goal THEN 'win'
                                           WHEN m.home_team_goal < m.away_team_goal THEN 'lose'
                                           ELSE 'draw'
                                       END AS result
                                FROM Match m
                                JOIN Team t ON m.home_team_api_id = t.team_api_id
                                UNION ALL
                                SELECT m.away_team_api_id AS team_api_id, t.team_long_name,
                                       CASE
                                           WHEN m.home_team_goal < m.away_team_goal THEN 'win'
                                           WHEN m.home_team_goal > m.away_team_goal THEN 'lose'
                                           ELSE 'draw'
                                       END AS result
                                FROM Match m
                                JOIN Team t ON m.away_team_api_id = t.team_api_id
                        ) AS combined_results
                        GROUP BY team_api_id, team_long_name
                        ORDER BY wins DESC
                        LIMIT 10
                        ;  """, conn)
q14
```

## Câu hỏi 19: Liệt kê thông tin cầu thủ:

- Quy đổi cân nặng sang kilogram
- Quy đổi chiều cao sang mét
- Tính chỉ số bmi = ( (weight* 0.453592) / (height/100)^2)
- Tính tuổi của cầu thủ

```python
# ĐIỀN CODE VÀO ĐÂY
q19 = pd.read_sql(""" SELECT p.id, 
                            p.player_api_id, 
                            p.player_name, 
                            p.player_fifa_api_id, 
                            p.birthday, 
                            pa.date,
                            STRFTIME('%Y', pa.date) - STRFTIME('%Y', p.birthday) AS Age,
                            ROUND(p.height / 100,2) AS height_m,
                            ROUND(p.weight * 0.453592,2) AS weight_kg,
                            ROUND(((p.weight* 0.453592) / (p.height/100)*2),2) AS bmi
                        FROM Player p
                        JOIN Player_Attributes pa
                        ON p.player_api_id = pa.player_api_id
                        ;  """, conn)
q19
```

## Câu hỏi 20: Liệt kê cầu thủ lớn tuổi nhất

```python
# ĐIỀN CODE VÀO ĐÂY
q20 = pd.read_sql(""" WITH Maxage AS(
                    SELECT MAX(STRFTIME('%Y', 'now') - STRFTIME('%Y', birthday)) AS Maxage
                    FROM Player
                    )
                     SELECT p.id, 
                            p.player_api_id, 
                            p.player_name, 
                            p.player_fifa_api_id, 
                            p.birthday,
                            STRFTIME('%Y', 'now') - STRFTIME('%Y', p.birthday) AS Age
                      FROM Player p
                      JOIN Maxage ma
                      ON (STRFTIME('%Y', 'now') - STRFTIME('%Y', p.birthday)) = ma.Maxage
                        ;  """, conn)
q20
```

## Câu hỏi 21: Liệt kê  cầu thủ có số trận đấu nhiều nhất
